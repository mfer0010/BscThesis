# Note: When working with totals we need to take into account the missing points

Hence its best, in my opinion to sample the columns, so we only use those that have consisten readings. Columns 1 and 42-61 have >100 non null entries, I'll use these.

```python
cols_to_drop = data.iloc[:, 2:41] #range(2,41)
df = data.drop(cols_to_drop,axis=1)
df = df.dropna()
df.reset_index()
df.info()
```

```python
df['Sum'] = df.drop(0,axis=1).sum(axis=1)
df.head()
```

```python
df = df.rename(columns={0:'Time Stamp'})
df.head()
```

```python
sample = df[:20]
plt.figure()
plt.plot(sample['Time Stamp'],sample['Sum'],'x')
plt.yscale('log')
plt.xlabel('Time Stamp')
plt.ylabel('Total Beam Loss over TL (Gy/s) (Log Scale)')
plt.title('Beam Losses of first 100 readings over TL')
plt.show()
```
