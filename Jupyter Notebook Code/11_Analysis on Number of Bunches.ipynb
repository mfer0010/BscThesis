```python
%reset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Required by Jupyter for inlining images
%matplotlib inline
```

```python
data = pd.read_csv('../Data/LHC_bunches_B1.txt',sep=' ', header=None)
data.head()
```

```python
data = data.rename(columns={0:'Time Stamp', 1:'Number of Bunches'})
```

Calculate Change in number of Bunches

```python
change = []
change.append(data['Number of Bunches'][0])
for i in range(1,len(data['Number of Bunches'])):
    change.append(data['Number of Bunches'][i]-data['Number of Bunches'][i-1])
data ['Change'] = change
```

```python
fig = plt.figure()
plt.plot(data['Time Stamp'][:100],data['Number of Bunches'][:100])
plt.xlabel('Time Stamp')
plt.ylabel('Number of Bunches')
plt.title('Time Series of Number of Bunches in the LHC')
```

```python
fig = plt.figure()
plt.plot(data['Time Stamp'],data['Change'],'x')
plt.xlabel('Time Stamp')
plt.ylabel('Change in Number of Bunches')
plt.title('Time Series of Change in Number of Bunches in the LHC')
```

As expected, the number of bunches increase with each injection
